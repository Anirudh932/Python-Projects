# Group By

The groupby() function in pandas is used to group data based on one or more columns in a DataFrame

```python
Amazon_Data.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
palette = sns.color_palette("Dark2")  # Choose your preferred palette

S_Qty=Amazon_Data.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.barplot(x='Size', y='Qty', data=S_Qty, palette=palette, edgecolor='black')
```

Note: From above Graph we can see that most of the people buys M-Size in the sales

```python
# Checking B2B Data  by using pie chart 
B2B_Check = Amazon_Data['B2B'].value_counts()
#  Plot the pie chart
colors = ['coral', 'cornflowerblue']  # Adjust colors as needed
plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%', colors=colors)
#plt.axis('equal')
plt.show()
```

```python
# Checking B2B Data  by using pie chart 
B2B_Check = Amazon_Data['B2B'].value_counts()
#  Plot the pie chart
colors = ['coral', 'cornflowerblue']  # Adjust colors as needed
plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%', colors=colors)
#plt.axis('equal')
plt.show()
```

Note : From above chart we can see that maximum i.e. 99.2% of buyers are from retailers and 0.8% are from B2B buyers.
    (Here False is Retailers and True is For B2B)

```python
# Prepare data for scatter plot
x_data = Amazon_Data['Category']  
y_data = Amazon_Data['Size'] 
# Plot the scatter plot

plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.scatter(x_data, y_data, color='Navy');
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize=(18, 6),facecolor='lightgray')
sns.countplot(data=Amazon_Data, x='ship-state', palette='vlag', edgecolor='black')

plt.xlabel('Ship-State', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state = Amazon_Data['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6),facecolor='lightgray')
sns.countplot(data=Amazon_Data[Amazon_Data['ship-state'].isin(top_10_state.index)], x='ship-state', palette='flare', edgecolor='black')
plt.xlabel('Ship-State', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.title('Distribution of  State', fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

Note: From above Graph we can see that most of the buyers are from Maharashtra.
